```python
import requests
from bs4 import BeautifulSoup

def top10k(n):
    rq = requests.get(f"https://tracker.gg/valorant/leaderboards/ranked/all/default?page={n}&region=global&act=4401f9fd-4170-2e4c-4bc3-f3b4d7d150d1")

    soup = BeautifulSoup(rq.text )

    pseudo = soup.findAll('span', class_='trn-ign__username')
    Hastag = soup.findAll('span', class_='trn-ign__discriminator')
    pseudo_Hastag = []
    #Hastag sans #:
    Hastag1 = [i.text.strip() for i in Hastag]
    Hastag_sans_hastag =[]
    for i in Hastag1:
        Hastag_sans_hastag.append(i.strip("#"))
    for i in range(len(pseudo)):
        pseudo_Hastag.append([pseudo[i].text.strip(),Hastag_sans_hastag[i]])
    return pseudo_Hastag
top10k(1)
```

```python
def KDA_player(pseudo,Hastag):

    url = f'https://tracker.gg/valorant/profile/riot/{pseudo}%23{Hastag}/weapons'
    pistol = None
    low = None
    Semi = None
    full = None


    response = requests.get(url)


    soup = BeautifulSoup(response.text, 'html.parser')
    if len(soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active'))>=3:
        pistol = soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active')[0].text

    if len(soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active')) >=3:
        low = soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active')[2].text

    if len(soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active')) >=3:
        Semi =  soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active')[1].text

    if len(soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active st-content__item-value--highlight'))>=1:
        full =soup.findAll('div', class_='st__item st-content__item-value st-content__item-value--active st-content__item-value--highlight')[0].text
    return [pseudo,Hastag,pistol,low,Semi,full]

def verification(pseudo,hastag):
    url = f"https://tracker.gg/valorant/profile/riot/{pseudo}%23{hastag}/weapons"
    rq = requests.get(url)
    if rq.status_code == 200:
        return 1
    else:
        return 0
```
