# Recuperation des pseudos et des Hastags


```python
import requests
from bs4 import BeautifulSoup

def top10k(n):
    rq = requests.get(f"https://tracker.gg/valorant/leaderboards/ranked/all/default?page={n}&region=global&act=4401f9fd-4170-2e4c-4bc3-f3b4d7d150d1")

    soup = BeautifulSoup(rq.text )

    pseudo = soup.findAll('span', class_='trn-ign__username')
    Hastag = soup.findAll('span', class_='trn-ign__discriminator')
    pseudo_Hastag = []
    #Hastag sans #:
    Hastag1 = [i.text.strip() for i in Hastag]
    Hastag_sans_hastag =[]
    for i in Hastag1:
        Hastag_sans_hastag.append(i.strip("#"))
    for i in range(len(pseudo)):
        pseudo_Hastag.append([pseudo[i].text.strip(),Hastag_sans_hastag[i]])
    return pseudo_Hastag
```

# Recuperation des Information D'un joueur

```python
import time



def info_Dunjoueur(pseudo,hastag):
    url = f"https://tracker.gg/valorant/profile/riot/{pseudo}%23{hastag}/overview"
    rq = requests.get(url)
    soup = BeautifulSoup(rq.text)
    info =soup.findAll('span', class_='value')
    info_joueur = []
    if len(info) >= 3:
        info.pop(0)
        info.pop(0)
        info.pop(0)

    info_joueur.append(pseudo)
    info_joueur.append(hastag)
    for i in range(len(info)):

        info_joueur.append(info[i].text.strip())
    return info_joueur


def verification(pseudo,hastag):
    url = f"https://tracker.gg/valorant/profile/riot/{pseudo}%23{hastag}/overview"
    rq = requests.get(url)
    if rq.status_code == 200:
        return 1
    else:
        return 0
```
